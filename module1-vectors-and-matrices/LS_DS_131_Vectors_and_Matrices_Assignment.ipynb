## 1.2 Create a three-dimensional vecor and plot it on a graph

```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

vectors=np.array( [ [0,0,1,1,-2,0], [0,0,2,1,1,0],[0,0,3,2,1,0],[0,0,4,0.5,0.7,0]]) 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for vector in vectors:
    v = np.array([vector[3],vector[4],vector[5]])
    vlength=np.linalg.norm(v)
    ax.quiver(vector[0],vector[1],vector[2],vector[3],vector[4],vector[5],
            pivot='tail',length=vlength,arrow_length_ratio=0.3/vlength)
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax.set_zlim([0,4])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
from math import e, pi
print(e)
print(pi)
```

```
#Scaling by 1.1, e & pi


#Move the scalars into their own vector
scale_factor_vector = np.array([1,1.1,e,pi])

#loop through each scalar
for scale_factor in scale_factor_vector:

    #Create a separate plot for each scalar
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    #Loop through the vectors that were used before
    for vector in vectors:

        #Scale and use the quiver to plot
        v = scale_factor*np.array([vector[3],vector[4],vector[5]])
        vlength=np.linalg.norm(v)
        ax.quiver(vector[0],vector[1],vector[2],vector[3],vector[4],vector[5],
                pivot='tail',length=vlength,arrow_length_ratio=0.3/vlength)
       
    #Set the plot size for each scalar so the arrows don't go off the grid
    range=4*scale_factor

    #Plot!
    ax.set_xlim([-range,range])
    ax.set_ylim([-range,range])
    ax.set_zlim([0,range])
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    plt.show()
       



```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\

\end{bmatrix}
\end{align}


```
H=np.array([[12,11],[7,10]])
J=np.array([[0,1,2],[7,10,4],[3,2,0]])
print(np.linalg.det(H))
print(np.linalg.det(J))
```

```
Hinv=np.linalg.inv(H)
Jinv=np.linalg.inv(J)
print(Hinv)
print(Jinv)
```
